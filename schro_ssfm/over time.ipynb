```python
from numpy import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV data
df = pd.read_csv("real1d-dyn.txt", sep='\\s+', header=None)
df.columns = ['x', 't', 'u']  # Assuming columns represent x, t, and u respectively

# Extract unique x and t values
x_unique = np.unique(df['x'].values)
t_unique = np.unique(df['t'].values)

# Step 2: Reshape u based on unique x and t values
u = df['u'].values
u_reshaped = u.reshape(len(t_unique), len(x_unique))  # Reshape u for easier plotting

# Step 3: Subsample the time steps (modify the step size to control number of lines)
subsample_rate = 10  # Change this value as needed to reduce the number of lines

plt.figure(figsize=(4, 6))

# Loop over every nth time step (subsampling)
for i in range(0, len(t_unique), subsample_rate):  # Adjust subsample_rate for more/fewer lines
    plt.plot(x_unique, u_reshaped[i, :], label=f't={t_unique[i]:.2f}')

# Step 4: Customize the plot
#plt.xticks(np.linspace(0, 2*pi, 5), ['0', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])

plt.xlabel('x')
plt.ylabel('ABS(psi(x))')
plt.title('h.o solution over time')
plt.legend(loc='upper right', title='Time')
plt.grid(True)

# Display the plot
plt.savefig("h.o solution over time.png")
plt.show()
```
